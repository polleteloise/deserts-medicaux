### Préparation de la table des codes NAF

```python
#copie colle la table + enregistre dans un excel
naf_df = pd.read_clipboard() 
naf_df.to_excel(os.path.join(DATA_DIR, NAF_REF_FILENAME), index = False)
```

```python
%%dpaste
pita3['NAF'] = np.where(pita3['etab_ape_l1'] == '86', 
                                pita3['activiteprincipaleetablissement'], 
                                pita3['activiteprincipaleunitelegale'])
```

```python
#pita3.head(100000)
```

```python
pita4 = pita3.merge(naf_df, on='NAF', how='inner') #on merge avec la table référence
```

```python
#pita4.head(100)
```

```python
#Autre méthode 1
#ps_df.loc[ps_df.cross_l1 == '86_86', 'code_activite'] = ps_df.activiteprincipaleetablissement
#ps_df.loc[ps_df.el_ape_l1 != "86", 'code_activite'] = ps_df.activiteprincipaleetablissement
#ps_df.loc[~(ps_df.activiteprincipaleetablissement.str.startswith(ACTIVITY_CODE)), 'code_activite'] = ps_df.activiteprincipaleunitelegale
```

```python
#pita4.drop(['unique_code'], axis=1)
```

```python
#Autre méthode 2
pita3['code_activite'] = pita3.apply(lambda x: x.activiteprincipaleetablissement
                                    if x.etab_ape_l1 == "86"
                                    else x.activiteprincipaleunitelegale,
                                    axis=1
                                    ) # axis = 1 signifie colonne, sinn c'est 0
```

```python
#Autre méthode 3, on crée une nouvelle colonne et on corrige après
pita3['code_activite']= pita3.activiteprincipaleetablissement
pita3['code_activite'] = pita3.where(pita3.etab_ape_l1 != 86,
                                    pita3.activiteprincipaleunitelegale,
                                     axis=1
                                    )
```

```python
#Transposition de la table et rassemblement des types de catégorie par communes
( pita3
 .merge(naf_df, on='NAF', how='inner')
 [['codecommuneetablissement', 'Catégorie']]
 .dropna(subset=['codecommuneetablissement'])
 .assign(count=1)
 .groupby(['codecommuneetablissement','Catégorie'])
 .sum()
 .reset_index()
 .pivot(index='codecommuneetablissement', columns='Catégorie', values='count')
 #.unstack(level=1)
 .fillna(0)
).to_csv(os.path.join(PROCESSED_DIR, PS_CITIES_FILENAME), sep=";")
```
